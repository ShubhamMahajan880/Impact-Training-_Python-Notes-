# **Lambda function**

* If the function doesn't have any name then its called as Lembda Funciton
Lemnda is called as High order Function

```python
#Addition
s = lambda a,b:a+b
r = s(10,5)
print(r)
```

```python
#Even numbers
s = lambda x:x%2==0
l = [1,2,3,4,5,6]
r = [p for p in l if(s(p))]
print(r)
```

```python
#Odd Numbers
s = lambda x:x%2!=0
l = []
for p in range(30, 50):
  l.append(p)
r = [i for i in l if(s(p))]
print(r)
```

**Qun :- Write a program to perform addition, subtraction, multiplication and division using one function and lambda.**

```python
def op(a,b,s):
  r = s(a,b)
  return r
s1 = op(10,5,lambda a,b:a+b)
s2 = op(10,5,lambda a,b:a-b)
s3 = op(10,5,lambda a,b:a*b)
s4 = op(10,5,lambda a,b:a//b)
print(s1,s2,s3,s4)
```

# Filter function

```python
# with filter without lambda

def even(n):
  return n%2==0
l = list(range(10,31))
r = list(filter(even, l))
print(r)
```

```python
# with filter with lambda

l = list(range(10,31))
r = list(filter(lambda x:x%2==0,l))
print(r)
```

# Positive and negative

```python
# with filter without lambda

def pos(n):
  return n>0
l = list(range(-5, 6))
r = list(filter(pos, l))
print(r)
```

```python
# with filter with lambda

l = list(range(-5,6))
r = list(filter(lambda x:x<0,l))
print(r)
```

**Qun :- Using filter and lambda return only upper case strings.**

```python
l = ['abc', 'ABC', 'xyz', 'WRT']
r = list(filter(lambda x:x.isupper(),l))
print(r)
```

# **Map Function()**


* Filter Requires the condition but map() requires a condition

# square

```python
#without lambda
def sq(n):
  return n*n
l = [1,2,3,4,5]
r = tuple(map(sq,l))
print(r)

#with lambda
l = [1,2,3,4,5]
r = tuple(map(lambda x:x*x, l))
print(r)
```

# **Reduce() function**

* Recude is a function and it is imported from funcy tools model

```python
# Addition of all numbers in a list

from functools import reduce
s = [1,2,3,4,5]
r = reduce(lambda a,b:a+b, s)
print(r)
```

```python
# Multiplication of all numbers in a list

from functools import reduce
s = [1,2,3,4,5]
r = reduce(lambda a,b:a*b, s)
print(r)
```

**Qun :- Find max and min using reduce and lambda function**

```python
from functools import reduce
d = [12,3,15,9,17,21,25,1]
max = reduce(lambda a,b:a if(a>b) else b, d)
print("max: ",max)
min = reduce(lambda a,b:a if(a<b) else b, d)
print("min: ",min)
```
