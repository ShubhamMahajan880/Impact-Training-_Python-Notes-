Qun :- John asked this time to smith write a logic to frequency count of each digit from the given number.

input: 2334

output:-

2 = 1

3 = 2

4 = 1 **bold text**

````python
num = input()
r = set(num)
l = list(r)
l.sort()

for i in l:
  print(i, "=",num.count(i))
````
