# Tuple :-
1) Tuple Objects can be created by using paranthesis & by calling functions

2) By assigning more than one value to single variable Tuple Object will be created

3) Tuple is immutable object, insertion order is preserved

4) Duplicate Elements are allowed

5) Tuple supports both Indexing & Slicing

6) Tuple is a collection of Homogeneous & Hetrogeneous Elements

```python
t = ()
type(t)
```

```python
t = tuple()
print(type(t))
```

```python
t = tuple(range(1,11))
t
```

```python
t = 1,2,3,4,5
print(t)
print(type(t))
```

```python
t = (1,2,3,4,5,1,2)
t
```

```python
t[2] #indexing is allowed but we can't update the values like, t[2] = 7
```
